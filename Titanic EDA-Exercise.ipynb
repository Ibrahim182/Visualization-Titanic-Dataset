Exploring titanic data set with seaborn
## The Data

We will be working with a famous titanic data set for these exercises. Later on in the Machine Learning section of the course, we will revisit this data, and use it to predict survival rates of passengers. For now, we'll just focus on the visualization of the data with seaborn:

```python
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df.info()
```

## making some visualization

## What age segment most of the passengers belong to? (hist,kde)

```python
from matplotlib import pyplot as plt
fig, axes = plt.subplots(1, 2, figsize=(15, 5), sharey=True)

sns.kdeplot(df['age'],ax=axes[0])

sns.distplot(df['age'],ax=axes[1]);
```

### What is the distribution of pfare?

```python
sns.distplot(df['fare'],bins=30);
```

## Describe the relationship between fare and age

```python
sns.jointplot('age','fare',data=df,space=0.5,kind='scatter');
```

### How many of the passengers did survive?

```python
sns.countplot(df['survived'], palette='viridis');
```

### How many of passengers of each class is alive?

```python
sns.countplot(df['alive'],hue=df['class'], palette='viridis');
```

### Which gender is reperesented more for each class?

```python
sns.countplot(df['class'],hue=df['sex'], palette='viridis');
```

### Which category of people is reperesented more for each class?

```python
sns.countplot(df['who'],hue=df['class'], palette='Set3');
```

### For each category of people get the count of alive people

```python
sns.countplot(df['who'],hue=df['alive'], palette='Set3');
```

### For each clas how many passengers are alone?

```python
sns.countplot(df['class'],hue=df['alone'], palette='pastel');
```

### How many passengers rode the ship for each embark town?

```python
sns.countplot(df['embark_town'],palette='Set3');
```

### How many passengers rode the ship for each embark town and still alive?

```python
sns.countplot(df['embark_town'],hue=df['alive'], palette='pastel');
```

### For each class, How many passengers rode the ship for each embark town ?

```python
sns.countplot(df['embark_town'],hue=df['class'], palette='Set3');
```

### What is the distribution of age for people who survived?

```python
sns.boxplot('survived','age',data=df,palette='pastel');
```

### What is the distribution of age for each class?

```python
sns.boxplot('class','age',data=df,palette='Paired');
```

### What is the distribution of age for each class and for each state of survival?

```python
sns.boxplot('class','age',hue='survived',data=df,palette='Paired');
```

### what is the sum of fare for each class for each embark town?

```python
sns.barplot(x='embark_town',y='fare',hue='class',data=df,estimator=np.sum,palette='Paired');
```

### Get the heat map of correlations of the dataframe

```python
c = df.corr()
sns.heatmap(c,cmap='viridis',linecolor='k',annot=True);
```
